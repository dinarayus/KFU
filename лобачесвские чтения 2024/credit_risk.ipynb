```python
x=data[["Annual_Income", "Monthly_Inhand_Salary",
        "Num_Bank_Accounts", "Num_Credit_Card", "Interest_Rate", "Num_of_Loan",
        "Delay_from_due_date", "Num_of_Delayed_Payment","Credit_Mix", "Outstanding_Debt", "Credit_History_Age", "Monthly_Balance"]]
y=data['Credit_Score']
```

```python
x['Credit_Mix'].unique()
x['Credit_Mix']=x['Credit_Mix'].replace({'Good': 2, 'Standard': 1, 'Bad': 0})
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(
    n_estimators=200,       # Number of trees in the forest
    max_depth=None,         # Maximum depth of the trees
    min_samples_split=2,    # Minimum number of samples required to split an internal node
    min_samples_leaf=1,     # Minimum number of samples required to be at a leaf node
    max_features='sqrt',     # Number of features to consider when looking for the best split
    random_state=42         # Seed for reproducibility
)
```

```python
rf_classifier.fit(x_train,y_train)
```

```python
pred=rf_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,pred)
print(f"accuracy of the model is {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Создание матрицы ошибок
cm = confusion_matrix(y_test, pred)

# Визуализация матрицы ошибок
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```
